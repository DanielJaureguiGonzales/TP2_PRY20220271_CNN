```python
from tensorflow.keras import datasets, layers, models, losses, Model, Input
from tensorflow.keras.applications.resnet_v2 import ResNet50V2
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Dropout,GlobalAveragePooling2D
```

#### NOTA: SE UTILIZA UNA ARQUITECTURA NEURONAL POR DEFECTO, PROX: RESTNET50

```python
model=Sequential()
#base_model = tf.keras.applications.resnet.ResNet152(weights = 'imagenet', include_top = False, input_shape = (224,224,3))
#for layer in base_model.layers:
#  layer.trainable = False
```

```python

model.add(Conv2D(16,(3,3),1, activation = 'relu', input_shape=(224,224,3),use_bias=True))
model.add(MaxPooling2D(pool_size=(2, 2)))


model.add(Conv2D(32,(3,3),1, activation = 'relu',use_bias=True))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(64,(3,3),1, activation = 'relu',use_bias=True))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(128,(3,3),1, activation = 'relu',use_bias=True))
model.add(MaxPooling2D(pool_size=(2, 2)))

         
model.add(Flatten())

model.add(Dense(256,activation='relu',use_bias=True,kernel_regularizer=tf.keras.regularizers.L2(0.01)))
model.add(Dropout(0.5))
model.add(Dense(128,activation='relu',use_bias=True,kernel_regularizer=tf.keras.regularizers.l2(0.01)))
model.add(Dropout(0.2))
model.add(Dense(4,activation='softmax',use_bias=True))

#x = Flatten()(base_model.output)
#x = Dense(1000, activation='relu')(x)
#predictions = Dense(4, activation = 'softmax')(x)
```

# ARQUITECTURA: ResNet50V2

```python
image_input = Input(shape=(224,224,3),batch_size=16)
model2 = ResNet50V2(input_tensor=image_input, include_top=False, weights='imagenet')


for layer in model2.layers:
    layer.trainable=False
    
model2.summary()

```

```python


x= GlobalAveragePooling2D()(model2.output)
x= Dense(512,activation='relu',name='fc1',kernel_regularizer='l2')(x)
x = Dropout(0.2)(x)
out = Dense(4,activation = 'softmax', name='output')(x)
custom_model=Model(image_input,out)


custom_model.summary()
```
